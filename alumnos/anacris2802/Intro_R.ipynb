```r
#Sacar el mejor d√≠a de la semana para volar
library(lubridate)
flights %>%
    mutate(week_day=wday(mdy(paste0(month,'-',day,'-',year)))) %>%
    group_by(week_day) %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
```

```r
#¬øHay alguna aerol√≠nea que tenga alg√∫n problema a nivel mes?

flights %>%
    group_by(carrier,month) %>%
    summarise(
        delay=mean(dep_delay,na.rm=TRUE )
    ) %>%
    arrange(desc(delay))
```

```r
#¬øHay alg√∫n avi√≥n problem√°tico?

flights %>%
    group_by(tailnum)  %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
    
```

```r
#¬øHay alg√∫n destino/origen problem√°tico?

flights %>%
    group_by(origin)  %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
```

```r
flights %>%
    group_by(dest)  %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
```

```r
#¬øHay alguna correlaci√≥n entre distancia y retrasos?
flights %>%
    group_by(distance)  %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
```
